## Введение

Когда говорят, что Python «поставляется с батарейками», имеют в виду именно его стандартную библиотеку. Установив интерпретатор, можно получить доступ к сотням готовых модулей — от работы с текстом и файлами до создания веб-серверов, многопоточности, математических вычислений, сжатия данных, криптографии, работы с электронной почтой. Всё это — без единого стороннего пакета, одной командой import.

Документация стандартной библиотеки содержит описание более 200 модулей.
https://docs.python.org/3/library/index.html

Обзор по группам модулей, от самых востребованных до узкоспециализированных, но одинаково полезных.

### Обработка текста
В эту категорию входят инструменты для работы со строками и текстовыми данными.
*   **`string`**: Предоставляет набор полезных строковых констант (например, `string.punctuation`) и классов для более гибкой работы с текстом, помимо встроенных методов.
*   **`re`**: Мощнейший модуль для **регулярных выражений**. Позволяет искать, заменять и извлекать сложные текстовые паттерны, что незаменимо при парсинге и валидации данных.
*   **`difflib`**: Помогает находить и вычислять сходства и различия между последовательностми (например, между двумя строками или файлами). Полезен для создания инструментов сравнения и поиска дубликатов.
*   **`textwrap`**: Управляет переносами строк в тексте. Может красиво отформатировать длинный текст для вывода в консоль или для вставки в другие документы.
*   **`unicodedata`**: Предоставляет доступ к огромной базе данных символов Юникода. Позволяет получать информацию о свойствах символа, например, является ли он буквой, цифрой или пробелом.

### Работа с типами данных
Здесь собраны модули, которые расширяют стандартные возможности по хранению и организации данных.
*   **`collections`**: Предлагает удобные и эффективные альтернативы стандартным контейнерам. Например, `deque` (двусторонняя очередь) для быстрых операций с обоих концов, `Counter` (счетчик) для подсчета хешируемых объектов и `defaultdict` (словарь по умолчанию).
*   **`enum`**: Позволяет создавать **перечисления** — наборы символических имен, привязанных к уникальным константам. Это делает код более читаемым и типобезопасным.
*   **`array`**: Предоставляет эффективный способ хранения большого количества однотипных числовых данных в компактном виде, экономя память по сравнению со списками.
*   **`struct`**: Позволяет преобразовывать данные между Python-объектами и **двоичными структурами** C. Необходим для работы с бинарными форматами файлов или сетевыми протоколами.
*   **`decimal`**: Модуль для высокоточных вещественных вычислений, который помогает избежать проблем с округлением, свойственных типу `float`. Особенно полезен в финансовой сфере.

### Математика и случайные числа
Модули для выполнения научных и статистических расчетов.
*   **`math`**: Содержит основные математические функции (тригонометрические, логарифмические, возведение в степень) и константы, такие как `pi` и `e`.
*   **`random`**: Предоставляет функции для генерации псевдослучайных чисел и выбора случайных элементов из последовательностей (например, `random.randint()`, `random.choice()`).
*   **`statistics`**: Включает функции для математической статистики: среднее арифметическое, медиану, моду, стандартное отклонение и другие.
*   **`cmath`**: Модуль для математических операций с **комплексными числами**.

### Функциональное программирование
Инструменты, вдохновленные парадигмой функционального программирования.
*   **`itertools`**: Содержит множество функций для создания и работы с **итераторами**. С его помощью можно создавать бесконечные циклы, комбинации и перестановки, что делает обработку последовательностей чрезвычайно эффективной.
*   **`functools`**: Включает функции для работы с другими функциями и вызываемыми объектами. Например, `partial` позволяет "заморозить" часть аргументов функции, создав новую, а `lru_cache` добавляет простое кеширование.
*   **`operator`**: Предоставляет функциональные аналоги стандартных операторов Python (`add`, `sub`, `itemgetter` и т.д.), которые удобно использовать с функциями высшего порядка вроде `map` и `filter`.

### Работа с файлами и операционной системой
Модули, которые позволяют вашему коду взаимодействовать с внешним миром.
*   **`os`**: Самый главный модуль для взаимодействия с операционной системой. Позволяет создавать, удалять и переименовывать файлы и директории, а также получать доступ к переменным окружения.
*   **`sys`**: Обеспечивает доступ к параметрам и функциям самого интерпретатора Python (например, аргументы командной строки, пути поиска модулей, стандартные потоки ввода/вывода).
*   **`pathlib`**: Предлагает **объектно-ориентированный** подход к работе с путями файловой системы. Это более современная и удобная альтернатива функциям из `os.path`.
*   **`shutil`**: Предоставляет высокоуровневые операции над файлами: копирование, перемещение, удаление целых директорий и управление дисковым пространством.
*   **`argparse`**: Незаменим для создания профессиональных консольных утилит. Позволяет легко парсить аргументы командной строки, генерировать подсказки и обрабатывать ошибки.
*   **`logging`**: Мощный и гибкий модуль для ведения журналов (логов) приложения. Позволяет настраивать уровни важности сообщений, направления их вывода (консоль, файл, сеть) и форматы.
*   **`io`**: Основной инструмент для работы с потоками ввода-вывода (файлы, строки, байтовые буферы). Лежит в основе многих других модулей.

### Сериализация и базы данных
Средства для сохранения данных между запусками программы.
*   **`json`**: Позволяет преобразовывать объекты Python в строки формата **JSON** (сериализация) и обратно (десериализация). Это стандартный формат обмена данными в веб-API.
*   **`pickle`**: Осуществляет сериализацию и десериализацию практически любых объектов Python в бинарный формат. Удобно, но **небезопасно** загружать данные из ненадежных источников.
*   **`shelve`**: Предоставляет постоянное хранилище для объектов Python, похожее на словарь. Фактически, это удобная надстройка над `pickle` для работы с файлами на диске.
*   **`csv`**: Предназначен для чтения и записи данных в формате CSV (Comma-Separated Values), популярном для обмена табличными данными.
*   **`sqlite3`**: Встраивает в Python мощную, легковесную **реляционную базу данных SQLite**. Позволяет работать с SQL-запросами без установки дополнительных серверов БД.

### Сжатие и архивирование
Модули для работы с популярными форматами сжатия и архивации.
*   **`zlib`**: Предоставляет функции для сжатия данных, совместимые с популярной библиотекой **gzip**.
*   **`gzip`**: Позволяет читать и записывать файлы в формате **gzip**, которые обычно имеют расширение `.gz`.
*   **`zipfile`**: Инструмент для создания, чтения и извлечения данных из **ZIP-архивов**.
*   **`tarfile`**: Предназначен для чтения и записи **TAR-архивов**, широко распространенных в Unix-системах.

### Криптография и безопасность
Инструменты для базовой защиты данных.
*   **`hashlib`**: Предоставляет криптографические хеш-функции (MD5, SHA-1, SHA-256 и др.), используемые для проверки целостности данных или хранения паролей.
*   **`hmac`**: Реализует механизм HMAC для проверки подлинности и целостности сообщений с использованием секретного ключа.
*   **`secrets`**: Специализированный модуль для **криптостойкой** генерации случайных чисел и токенов, безопасный для использования в приложениях.

### Дата, время и календари
Модули, без которых не обходится ни одно приложение, работающее со временем.
*   **`datetime`**: Ключевой модуль для работы с датой и временем. Предоставляет классы `date`, `time`, `datetime` и `timedelta` для выполнения различных операций и арифметики с временными интервалами.
*   **`calendar`**: Позволяет выполнять операции с календарем: выводить на печать календарь на месяц или год, определять день недели для заданной даты.
*   **`time`**: Обеспечивает доступ к функциям для работы со временем на уровне операционной системы (например, `time.sleep()` для приостановки выполнения программы).

### Многопоточность, процессы и асинхронность
Модули для параллельного выполнения кода и повышения производительности.
*   **`threading`**: Позволяет запускать несколько потоков в рамках одного процесса. Идеально подходит для I/O-связанных задач (ожидание ответа от сети, диска).
*   **`multiprocessing`**: Предназначен для выполнения кода в нескольких процессах. Решает проблему глобальной блокировки интерпретатора (GIL) и идеален для CPU-связанных задач (сложные вычисления).
*   **`concurrent.futures`**: Предоставляет высокоуровневый интерфейс для асинхронного выполнения задач с использованием пулов потоков или процессов.
*   **`subprocess`**: Позволяет запускать новые процессы, взаимодействовать с их потоками ввода/вывода и получать коды возврата. С его помощью можно выполнять системные команды.
*   **`asyncio`**: Библиотека для написания **асинхронного кода** с использованием ключевых слов `async` и `await`. Обеспечивает высокую производительность для приложений с большим количеством сетевых соединений.

### Сеть и интернет-протоколы
Модули для создания сетевых приложений и взаимодействия по протоколам Интернета.
*   **`socket`**: Низкоуровневый сетевой интерфейс. Позволяет создавать как клиентские, так и серверные приложения, используя сокеты Berkeley.
*   **`ssl`**: Добавляет уровень шифрования SSL/TLS к сетевым соединениям, что необходимо для безопасной передачи данных по протоколу HTTPS.
*   **`urllib`**: Предоставляет набор модулей для работы с URL-адресами, включая отправку запросов, парсинг URL и обработку ошибок протокола HTTP.
*   **`http.server`**: Позволяет легко запустить простой HTTP-сервер, что может быть полезно для тестирования или быстрого обмена файлами.
*   **`smtplib`**: Реализует клиентскую часть протокола SMTP для отправки электронной почты.
*   **`uuid`**: Генерирует уникальные идентификаторы (Universally Unique IDentifier), которые можно использовать для идентификации объектов в распределенных системах.

### Веб-технологии и форматы данных
Модули для обработки веб-контента и популярных форматов.
*   **`html`**: Содержит функции для экранирования и обработки HTML-сущностей, что полезно для защиты от XSS-атак.
*   **`xml`**: Пакет, содержащий несколько подмодулей для парсинга и обработки XML-документов (`xml.etree.ElementTree`, `xml.dom`, `xml.sax`).
*   **`json`**: (уже упоминался в сериализации) — ключевой формат для веб-API.
*   **`base64`**: Предоставляет функции для кодирования бинарных данных в текстовый формат Base64, широко используемый для передачи данных по электронной почте или в JSON.

### Разработка и отладка
Инструменты, которые помогают писать, тестировать и документировать код.
*   **`unittest`**: Фреймворк для создания и запуска автоматических тестов, вдохновленный библиотекой JUnit в Java.

*   **`typing`**: Предоставляет поддержку подсказок типов (type hints), что улучшает читаемость кода и позволяет использовать статические анализаторы для поиска ошибок.
*   **`venv`**: Инструмент для создания изолированных виртуальных окружений Python. Позволяет управлять зависимостями разных проектов независимо.
*   **`timeit`**: Простой в использовании инструмент для измерения времени выполнения небольших фрагментов кода[reference:50].

## csv

CSV (от англ. Comma-Separated Values — значения, разделённые запятыми) — текстовый формат, предназначенный для представления табличных данных. Строка таблицы соответствует строке текста, которая содержит одно или несколько полей, разделенных запятыми. 

В Python работа с CSV-файлами поддерживается стандартным модулем csv, предоставляющем следующие основные объекты и функции:
- `csv.reader(csvfile, dialect='excel', **fmtparams)` - создает и возвращает объект для чтения строк
- `csv.writer(csvfile, dialect='excel', **fmtparams)` - создает и возвращает объект для записи строк
- `class csv.DictReader(csvfile, fieldnames=None, restkey=None, restval=None, dialect='excel', *args, **kwds)` - создает и возвращает объект для чтения данных в словари
- `class csv.DictWriter(csvfile, fieldnames, restval='', extrasaction='raise', dialect='excel', *args, **kwds)` - создает и возвращает объект для записи данных из словарей

````python
import csv

with open('names.csv', 'w', newline='') as csvfile:
    fieldnames = ['first_name', 'last_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerow({'first_name': 'Baked', 'last_name': 'Beans'})
    writer.writerow({'first_name': 'Lovely', 'last_name': 'Spam'})
    writer.writerow({'first_name': 'Wonderful', 'last_name': 'Spam'})
````

````python
with open('names.csv', newline='') as csvfile:

    reader = csv.DictReader(csvfile)

    for row in reader:

        print(row)
````

````python
with open('names.csv', newline='') as csvfile:

    reader = csv.reader(csvfile)

    for row in reader:

        print(row)
````

## json

По сути, формат json является стандартом текстовой сериализации объектов.

Удобный, легковесный, человекочитаемый. 

JSON-текст представляет собой одну из двух структур:
- набор пар ключ: значение (словарь в терминологии Python), где ключ - строка, значение - любой тип;
- упорядоченный набор значений (список в терминологии Python).

Значением может быть:
- строка (в двойных кавычках);
- число;
- логическое значение (true/false);
- null;
- одна из структур.

Пример: 
```json
{
    "Модель": "ChinaPhone11",
    "Цена": {
        "Новый": 1999.9,
        "Б/у": 800,
        "На запчасти": 200
    },
    "Цвет": ["Красный", "Золотой"],
    "Комментарий": null,
    "Топ-бренд": true
}
```

JSON-формат поддерживается стандартным пакетом json, предоставляющем следующие основные функции:
- `json.dumps(obj, skipkeys=False, ensure_ascii=True, check_circular=True, allow_nan=True, cls=None, indent=None, separators=None, default=None, sort_keys=False, **kw)`
- `json.loads(s, encoding=None, cls=None, object_hook=None, parse_float=None, parse_int=None, parse_constant=None, object_pairs_hook=None, **kw)`



````python
import json

d = {
    "Модель": "ChinaPhone",
    "Цена": {
        "Новый": 1999.9,
        "Б/у": 800,
        "На запчасти": 200
    },
    "Цвет": ["Красный", "Золотой"],
    "Комментарий": None,
    "Топ-бренд": True,
    "Популярность": float('inf')
}

j = json.dumps(d, indent=2, ensure_ascii=False)  # indent + print сделают вывод json красивым
print(j)
````

````python
d = json.loads(j)
d
````

## пикл

Стандартный модуль pickle, при помощи которого можно сохранять практически любой объект Python в двоичном формате (файле), а затем извлекать его обратно

Основные функции модуля:
- `pickle.dump(obj, file, protocol=None, *, fix_imports=True)` - сериализация или пиклинг
- `pickle.load(file, *, fix_imports=True, encoding='ASCII', errors='strict')` - десериализация или анпиклинг

Pickle является основным способом сериализации объектов для python, является инструментом по-умолчанию для передачи данных между процессами python (при использовании многопроцессорности)

Следующие типы могут быть сериализованы с помощью pickle:

    None, True, False;
    int, float, complex;
    str, bytes, bytearrays;
    tuple, list, set, dict (если содержат только сериализуемые объекты);
    functions (встроенные и пользовательские), определённые на верхнем уровне модуля;
    class, определённые на верхнем уровне модуля;
    инстансы таких классов, результат вызова __getstate__() которых сериализуем.


Минусы формата pickle:
- специфичен для Python;
- небезопасен.


````python
import pickle
from dataclasses import dataclass

@dataclass
class User:
    name: str
    role: str
        
def plus(a, b):
    return a + b

data = {
    'a': [1, 2.0, 3+4j],
    'b': ("character string", b"byte string"),
    'c': {None, True, False},
    'd': User,
    'e': User('A', 'B'),
    'f': plus,
}

with open('data.pickle', 'wb') as f:
    pickle.dump(data, f, pickle.HIGHEST_PROTOCOL)
````

## xml

XML (от англ. eXtensible Markup Language) — «расширяемый язык разметки». Рекомендован Консорциумом Всемирной паутины (W3C). Спецификация XML описывает XML-документы. XML разрабатывался как язык с простым формальным синтаксисом, удобный для создания и обработки документов как программами, так и человеком, с акцентом на использование в Интернете. Язык называется расширяемым, поскольку он не фиксирует разметку, используемую в документах

Пример:
```xml
<?xml version="1.0" encoding="utf-8"?>
<!DOCTYPE recipe>
<recipe name="хлеб" preptime="5min" cooktime="180min">
   <title>
      Простой хлеб
   </title>
   <composition>
      <ingredient amount="3" unit="стакан">Мука</ingredient>
```

## random 

https://docs.python.org/3/library/random.html

Модуль для псевдо-случайных выборок с различными распределениями

В основе работы генератора псевдо-случайных чисел (ГПСЧ) лежит алгоритм под названием [Вихрь Мерсенна](https://ru.wikipedia.org/wiki/%D0%92%D0%B8%D1%85%D1%80%D1%8C_%D0%9C%D0%B5%D1%80%D1%81%D0%B5%D0%BD%D0%BD%D0%B0) 

Функции, которые позволяют зафиксировать состояние генератора ПСЧ
- `random.seed(a=None, version=2)` - инициализирует и фиксирует ГПСЧ значением `a`, если `a` пусто или `None` - текущем временем
- `random.getstate()` - возвращает текущее состояние ГПСЧ
- `random.setstate(state)` - задаёт текущее состояние ГПСЧ

````python
import random
random.seed(1)  # инициализирует и фиксирует ГПСЧ значением 1 (фиксация нужна для повторимости, например в тестах)
````

````python
state = random.getstate()
````

````python
random.setstate(state)
print(random.random())
print(random.random())
````

````python
random.seed()
random.random()
````

Целочисленные функции
- `random.randrange(stop)` - вернёт cлучайное число до `stop`
- `random.randrange(start, stop[, step])`
- `random.randint(a, b)` - псевдоним randrange(a, b+1)

````python
random.randrange(2, 10)
````

````python
random.randint(2, 10)
````

Функции для работы с последовательностями
- `random.choice(seq)` - вернёт 1 cлучайный элемент из `seq`
- `random.choices(population, weights=None, *, cum_weights=None, k=1)`
- `random.shuffle(x)` - на месте перемешивает последовательность `x`

````python
seq = [12, 17, 19, 300, 425]
random.choice(seq)
````

````python
random.choices(seq, [1, 10, 1, 1, 1], k=2)
````

````python
random.shuffle(seq)
seq
````
